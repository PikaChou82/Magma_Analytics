```python
# On ne garde que les titres
df =df.loc[:,['backdrop_path', 'overview', 'poster_path', 'original_language','vote_average', 'origin_country', 'parentTconst', 'title_series','genres', 'startYear']]
df = df.drop_duplicates()
df = df.reset_index()
```

```python
from sklearn.decomposition import PCA
from sklearn.metrics import pairwise_distances
from sklearn.neighbors import NearestNeighbors
# Import des librairies

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
import folium
from sklearn.model_selection import learning_curve
import missingno as msno
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import RobustScaler

def reco_knn(df_filtre):
    # SÃ©lection des colonnes numÃ©riques pertinentes pour la recommandation
    df_filtre['genres_'] = df_filtre['genres'].factorize()[0]
    colonnes_numeriques = ['vote_average', 'startYear', 'genres_']
    df_numerique = df_filtre[colonnes_numeriques].dropna()  # Supprimer les lignes contenant des NaN

    # Normalisation des donnÃ©es
    df_numerique_normalized = (df_numerique - df_numerique.mean()) / df_numerique.std()

    """ MÃ©thode KNN """
    knn = NearestNeighbors(n_neighbors=4, metric='euclidean')  # 3 voisins et la serie elle mÃªme Ã  exclure
    knn.fit(df_numerique_normalized)

    # Trouvons les indices des voisins
    distances, indices = knn.kneighbors(df_numerique_normalized)

    # Ajouter les sÃ©ries similaires dans le DataFrame
    for i, voisin_indices in enumerate(indices):
        df_filtre.loc[i, 'voisin1'] = df_filtre.iloc[voisin_indices[0]]['title_series']
        df_filtre.loc[i, 'voisin2'] = df_filtre.iloc[voisin_indices[1]]['title_series']
        df_filtre.loc[i, 'voisin3'] = df_filtre.iloc[voisin_indices[2]]['title_series']
        df_filtre.loc[i, 'voisin4'] = df_filtre.iloc[voisin_indices[3]]['title_series']

    return df_filtre

# Ajoutons la fonction recommandations au df filtrÃ©
df_filtre_reco = reco_knn(df)
```
