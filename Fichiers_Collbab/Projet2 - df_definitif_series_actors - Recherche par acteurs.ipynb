All_actors est un DataFrame issu d'un chunck du fichier name.basics.tsv et d'un explode du même fichier colonne['tconst'] dans un colab individuel.

DataFrame nettoyé et renommé ALL_ACTORS que l'on importe ci-dessous.

## IMPORT DE ALL_ACTORS (df name nettoyé)

```python
all_actors = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/PROJET 2 /all_actors.csv")
all_actors = all_actors.drop("Unnamed: 0", axis = 1)
all_actors
```

## MERGE AVEC LE DataFrame "df_definitif_series"

```python
# Import du fichier depuis github
url2 ="https://raw.githubusercontent.com/PikaChou82/Magma_Analytics/refs/heads/main/Datasets/df_definitif_series.csv"
series = pd.read_csv(url2, sep=",")
```

```python
series.head()
```

### Merge réalisé avec les colonnes id de chacun des fichiers. Il est ensuite stocké dans un DataFrame df.

```python
df = all_actors.merge(series, left_on='tconst', right_on='parentTconst')
df = df.drop("tconst_x", axis=1) # supprimer une colonne doublon
```

```python
# Visuel du df
df.head()
```

### Filtre sur df pour conserver les id de séries et les noms d'acteurs. + Remplacer les valeurs nulles.

```python
data = df[["parentTconst", "primaryName"]]
data
```

```python
# Y a t il des doublons
data.duplicated().sum()
```

```python
# Supprime les doublons
data5 = data.drop_duplicates()
data5
```

### Afin de pouvoir joindre les noms d'acteurs au DataFrame initial df_definitif_series, on regroupe tous les noms d'acteurs par parentTconst afin d'éviter les doublons des id series.

```python
data1 = data5.groupby("parentTconst")["primaryName"].apply(lambda x: "%s" % ', '.join(x))
data1
```

### Jointure entre le fichier series (df_definitif_series) et la data1.

```python
df_definitif_series_actors = pd.merge(series, data1, on="parentTconst", how="outer")
```

```python
df_definitif_series_actors.shape
```

```python
df_definitif_series_actors.info()
```

```python
# Vérification de valeur nulle
df_definitif_series_actors.isna().sum()
```

```python
# Remplace les valeurs nulles
df_definitif_series_actors.fillna({"backdrop_path": "None"}, inplace=True)
df_definitif_series_actors.fillna({"overview": "None"}, inplace=True)
df_definitif_series_actors.fillna({"poster_path": "None"}, inplace=True)
df_definitif_series_actors.fillna({"primaryName": "Unknown"}, inplace=True)
```

```python
# Vérification de valeur nulle
df_definitif_series_actors.isna().sum()
```
