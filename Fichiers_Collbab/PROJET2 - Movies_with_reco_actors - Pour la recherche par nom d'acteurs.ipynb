```python
import pandas as pd
```

# Remodelage du DataFrame movies_with_reco pour ajouter les acteurs.

Import de actors pour récupérer les noms des acteurs et de films (movie_with_reco)

- Nous souhaitions pouvoir intégrer un filtre sur les acteurs dans notre page de recherche afin de permettre à l'utilisateur de réaliser une recherche axée sur ce détail.

- Nous somme passé par différente étapes avant de conclure que le plus efficace serait tous simplement d'ajouter les noms d'acteurs au DataFrame movies_with_reco.


---
*Le DataFrame movies_with_reco quant à lui contient dans différentes colonnes les recommandations de films réalisées avec la méthode KNN en prenant en compte les paramètres suivants : années, genres, 3 acteurs principaux, pays de production & directeurs de production.*

---



## Visualisation de actors

- Observation du DataFrame actors pour identifier les colonnes à conserver et le nom de la colonne à utiliser pour effectuer la jointure.

```python
url="https://raw.githubusercontent.com/PikaChou82/Magma_Analytics/refs/heads/main/Datasets/actors.csv"

actors = pd.read_csv(url, sep=",")
```

```python
actors.head()
```

### Extraction des colonnes primaryName et tconst.

- Les colonnes qui nous intéressent sont tconst (id des films) et primaryName(nom d'acteurs).

```python
actors2 = actors.loc[:, ["primaryName", "tconst"]]
```

```python
actors2
```

## Import et visualisation de films (movies_with_reco)

- On observe que la colonne imdb_id nous servira de pillier pour la jointure avec tconst de actors.

```python
url1 = "https://raw.githubusercontent.com/PikaChou82/Magma_Analytics/refs/heads/main/Datasets/movies_with_reco.csv"
films = pd.read_csv(url1, sep=',')
```

```python
films.head()
```

## Jointure des deux datas par tconst et imdb_id.


- Jointure entre les deux dataframe par id de film tconst pour l'un et imdb_id pour l'autre. On conserve toutes les lignes et toutes les colonnes et on abandonne tconst qui fait doublon avec imbd_id.

- On le conserve dans le nouveau dataframe nommé data_actors2.

```python
dataset = pd.merge(films, actors2, left_on="imdb_id", right_on="tconst", how="outer")
data_actors2 = dataset.drop("tconst", axis=1)
data_actors2.head()
```

```python
data_actors2.to_csv("data_actors2.csv", index=False)
```

### Nouveau filtre afin de conserver les id de films et remplacer les Nan de primaryName par "Unknown".

```python
data = data_actors2[["imdb_id", "primaryName"]]
```

```python
data
```

```python
# On supprime les NaN de imdb_id et on remplace les NaN de primaryName par "Unknown"
data5 = data.dropna(subset="imdb_id")
data5["primaryName"].fillna("Unknown", inplace=True)
data5
```

### Afin de nous permettre de joindre les noms d'acteurs au DataFrame initial movie_with_reco, on regroupe tous les noms d'acteurs par id de films afin d'éviter les doublons des id films.

```python
data1 = data5.groupby("imdb_id")["primaryName"].apply(lambda x: "%s" % ', '.join(x))
```

```python
data1
```

```python
data1.info()
```

## Jointure entre le fichier films (movies_with_reco) et la data1 (correspondant au nettoyage de data_actors2).

```python
movies_with_reco_actors = pd.merge(films, data1, on="imdb_id", how="outer")
```

```python
movies_with_reco_actors.head()
```

### Enregistrement en CSV du fichier pour le streamlit.

```python
movies_with_reco_actors.to_csv("movies_with_reco_actors.csv", index=False)
```
