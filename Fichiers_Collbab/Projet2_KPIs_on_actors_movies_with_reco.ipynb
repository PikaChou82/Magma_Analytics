## MERGE

```python
actors_movies = actors.merge(movies_reco, left_on='tconst', right_on='imdb_id')
actors_movies
```

## CLEANING
->>>> actors_movies qui devient le fichier pour établir les KPIs

```python
# Y a t il des doublons.
actors_movies.duplicated().sum()
```

```python
# Existe t il des valeurs NaN
actors_movies.isna().sum()
```

```python
# On garde uniquement les acteurs
a = actors_movies[actors_movies['category']=="actor"]
a
```

```python
# On garde uniquement les actrices
b= actors_movies[actors_movies['category']=='actress']
b
```

```python
# Concaténation de actor et actress
actors_movies_reco = pd.concat([a, b], ignore_index=True)
actors_movies_reco
```

```python
# Enregistrement du fichier
actors_movies_reco.to_csv('/content/drive/MyDrive/Colab Notebooks/PROJET 2 /dataset2/actors_movies_reco.csv', index=False)
```

# KPIs

```python
actors_movies_reco.columns
```

```python
# Nombre total d'acteurs.
number_actors = actors_movies_reco['primaryName'].nunique()
number_actors
```

```python
# Top 10 des acteurs les plus actifs.
actors_movies_reco['primaryName'].value_counts().head(10)
```

```python
# Répartition des acteurs par genre de films : Nombre absolu d'acteurs dans chaque genre de films.
number_actors_genres = actors_movies_reco.groupby('genres')['nconst'].nunique().sort_values(ascending=False)
number_actors_genres
```

```python
actors_movies_reco.groupby('genres')['nconst'].nunique().sort_values(ascending=True).plot(kind='barh',
                                title="Nombre d'acteurs par genre de films") ;
```

```python
# Pourcentage d'acteurs dans chaque genre de films.
percent_num_act_genres = (number_actors_genres/number_actors)*100
percent_num_act_genres
```

```python
percent_num_act_genres.plot(kind='line', title="Acteurs par genre de films (%)", ylabel='Poucentage %', xlabel='GENRES')
```

```python
# Nombre de films par acteur par genre : Nombre de films par acteur, ventilé par genre, pour voir les préférences de genre des acteurs.
actors_movies_reco.groupby(['primaryName', 'genres'])['genres'].value_counts().sort_values(ascending=False).head(60)
```

```python
actors_movies_reco.groupby(['primaryName', 'genres'])['genres'].value_counts().sort_values(ascending=False).head(20).plot(kind='barh', title="Nombre de film par acteur par genre")
```

```python
# Nombre moyen d'acteurs par film.
number_actors_by_film = actors_movies_reco["title"].value_counts()
number_actors_by_film
```
